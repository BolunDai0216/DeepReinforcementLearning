```python
import numpy as np
import tensorflow as tf
import pandas as pd
from sklearn.model_selection import train_test_split
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv('breast-cancer-wisconsin.txt', header=None)
```

```python
feature = data.iloc[:, 1:10]
label = data.iloc[:, 10]
```

```python
label = tf.keras.utils.to_categorical(label.values/2-1, num_classes=2, dtype='float32')
```

```python
X_train, X_val, y_train, y_val = train_test_split(feature, label, test_size=0.2, random_state=3)
```

```python
model = Sequential([
    Dense(300, activation='relu', input_shape=(9,)),
    Dense(300, activation='relu'),
    Dense(2, activation='softmax')
])

model.compile(
    optimizer='Adam',
    loss='binary_crossentropy',
    metrics=['mae']
)

model.summary()
```

```python
history = model.fit(X_train, y_train, epochs=300, batch_size=64)
```

```python
loss = history.history['loss']
val_mae = history.history['val_mae']
```

```python
plt.figure()
plt.plot(loss, label='Training Loss')
plt.plot(val_mae, label='Validation MAE')
plt.legend()
plt.xlabel('Epochs')
plt.savefig('training_curve.png', dpi=400)
```

```python
y_val_predict = model.predict(X_val)
error = np.argmax(y_val_predict, axis=1)-np.argmax(y_val, axis=1)
```
